```python
# iloc = index-based selection

# --> df.iloc[0] --> this would give the first row as Series 
                   # Series (Indexing: Has an index associated with the values, which can be integer-based or label-based (similar to a dictionary))

# when using iloc for slicing, the 'stop' index is EXCLUSIVE
# --> df.iloc[0:3] --> this would give the 1st 3 rows exclusing the 3rd index


# --> df.iloc[0:] == df.iloc[:] --> selects the 0th row to all the rows --> selects all rows from a DataFrame starting from index 0 to the end
# effectively returns the entire DataFrame.

# --> df.iloc[:0] == df.iloc[0:0] --> starts from the 0th row and ends before 0th row --> returns Empty


# can use ',' to get specific elements
# --> df.iloc[0,1] --> element returned at row 0, col 1

# --> df.iloc[:, 0] --> all rows of 1st column
```

```python
# loc = Label-based selection

# --> df.loc[0] --> returns the 1st row and all columns values

# --> df.loc[0:3, :] --> returns 0-3 INCLUSIVE rows and all cols

# --> df.loc[0:0] --> returns the 1st row and all columns

# --> df.loc[:, 'ph'] --> returns all rows of the column named "ph"

# --> df.loc[5, 'color'] --> returns the 5th row label value of the column 'color'
```

```python
# to turn HTML from the WWW into useful data, we need to parse it and extract structural data 

# This is why BeautifulSoup exists
# --> it makes tag soup beautiful and helps extract info

# also very important method = prettify
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
# url
url = 'https://www.python.org/~guido/'

# send the request
r = requests.get(url)

# Extracts the response as html
html_doc = r.text
```

```python
# Create a BeautifulSoup object from the HTML
soup = BeautifulSoup(html_doc)

# Prettify the BeautifulSoup object
pretty_soup = soup.prettify()

print(pretty_soup)
```

```python
# figure out how to extract the text from the webpage, along with printing the webpage's title.
```

```python
# fro the title --> you only have the title attribute
guido_title = soup.title

# Print the title of Guido's webpage to the shell
print(guido_title)
```

```python
# for extracting just the text, you have get method
guido_text = soup.get_text()


print(guido_text)
```

```python
# getting the hyperlinks

# Find all 'a' tags (which define hyperlinks), you will pass <a> element from the HTML w/o the angle brackets and as a character 
a_tags = soup.find_all('a')
# find_all is a function that lets us find the things we want in the HTML

# Print the URLs to the shell
for any_link in a_tags:
    print(any_link.get('href'))
```
